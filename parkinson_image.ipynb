```python
import matplotlib.pyplot as plt
import numpy as np

# Define points
A = (0, 8)
B = (36/23, 88/23)
C = (4, 2)
D = (12, 0)

# Create plot
plt.figure(figsize=(8, 6))

# Define x values for plotting
x_vals = np.linspace(0, 14, 100)

# Define line equations based on the points
# Line AB (using slope between A and B)
y_AB = (88/23 - 8) / (36/23 - 0) * (x_vals - 0) + 8

# Line BC (using slope between B and C)
y_BC = (2 - 88/23) / (4 - 36/23) * (x_vals - 36/23) + 88/23

# Line CD (using slope between C and D)
y_CD = (0 - 2) / (12 - 4) * (x_vals - 4) + 2

# Plot lines
plt.plot(x_vals, y_AB, color='brown', label="AB")
plt.plot(x_vals, y_BC, color='blue', label="BC")
plt.plot(x_vals, y_CD, color='gray', label="CD")

# Plot points A, B, C, and D
plt.scatter(*A, color='black', zorder=5)
plt.scatter(*B, color='black', zorder=5)
plt.scatter(*C, color='black', zorder=5)
plt.scatter(*D, color='black', zorder=5)

# Annotate points
plt.text(A[0], A[1], 'A(0,8)', verticalalignment='bottom', horizontalalignment='right')
plt.text(B[0], B[1], 'B(36/23,88/23)', verticalalignment='bottom', horizontalalignment='right')
plt.text(C[0], C[1], 'C(4,2)', verticalalignment='top', horizontalalignment='right')
plt.text(D[0], D[1], 'D(12,0)', verticalalignment='top', horizontalalignment='right')

# Shading the feasible region between lines AB and BC, then BC and CD
plt.fill_between(x_vals, y_AB, y_BC, where=(x_vals >= A[0]) & (x_vals <= C[0]), color='yellow', alpha=0.3)
plt.fill_between(x_vals, y_BC, y_CD, where=(x_vals >= C[0]) & (x_vals <= D[0]), color='yellow', alpha=0.3)

# Labels and grid
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlim(0, 14)
plt.ylim(0, 9)

plt.legend()
plt.show()





```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definisikan rentang x untuk plotting, dari 0 hingga 15
x_values = np.linspace(0, 15, 100)

# Persamaan garis berdasarkan hasil sebelumnya
y1 = -0.75 * x_values + 5
y2 = -2.67 * x_values + 8
y3 = -0.25 * x_values + 3

# Titik baru dengan label A, B, C, dan D beserta koordinat tanpa angka desimal
new_points = {"A": (0, 8), "B": (36/23, 88/23), "C": (4, 2), "D": (12, 0)}
x_new_points, y_new_points = zip(*new_points.values())

# Plot garis
plt.figure(figsize=(10, 6))
plt.plot(x_values, y1, label="y = -0.75x + 5", color="blue")
plt.plot(x_values, y2, label="y = -2.67x + 8", color="green")
plt.plot(x_values, y3, label="y = -0.25x + 3", color="red")

# Plot titik A, B, C, D
plt.scatter(x_new_points, y_new_points, color="black", marker="o")

# Label titik dengan offset lebih besar
labels = {
    "A": "A(0,8)",
    "B": "B(36/23,88/23)",
    "C": "C(4,2)",
    "D": "D(12,0)"
}
offsets = {
    "A": (-0.5, 1.0),
    "B": (1.0, -1.0),
    "C": (1.0, -1.0),
    "D": (1.0, -1.0)
}

for label, (x, y) in new_points.items():
    dx, dy = offsets[label]  # Offset untuk menghindari tumpang tindih dengan garis
    plt.text(x + dx, y + dy, labels[label], fontsize=10, ha="right" if dx > 0 else "left", color="black", weight="bold")

# Batas sumbu x dan y dari 0 hingga 15
plt.xlim(0, 15)
plt.ylim(0, 15)

# Menambah label sumbu, judul, dan legenda
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lines through Points A, B, C, D with Enhanced Label Positions")
plt.legend()
plt.grid(True)

# Tampilkan plot
plt.show()
```
